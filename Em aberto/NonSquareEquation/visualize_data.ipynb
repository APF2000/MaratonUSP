```python
import numpy as np
import matplotlib.pyplot as plt

import os
import imageio
```

```python
def s(x):
    base = 10
    soma = 0

    while(x != 0):
        soma += ( x % base )
        x //= base
    
    return soma

def formula(x):
    return ( x + s(x) ) * x
```

```python
def choose_color(n):
    n = n % 3
    colors = { 0: (1, 0, 0), 1: (0, 1, 0), 2: (0, 0, 1) }
    return colors[n]


def create_frame(passo, n_min, n_max, fig_path = 'TempImages/frame.svg'):
    #print(f'n_min = {n_min}, n_max={n_max}')
    X = np.array([i for i in range(n_min, n_max)])#np.linspace(n_min, n_max, passo)
    y = np.array([formula(el) for el in X])
    #print('\n-----Creating frame-----')
    #print(X, len(X))
    #print(y, len(y))
    X_mult_10 = [el for el in X if el % 10 == 0]
    print('x_mult10: ', X_mult_10, len(X_mult_10))

    for n in range(len(X_mult_10)):
        x_hline = X_mult_10[n]
        y_hline = formula(x_hline)
        color = choose_color(x_hline // 10)
        
        n = (n * 10)
        #print('n = ', n, formula(n), y[n])
        #plt.hlines(y = formula(n), xmin =  n, xmax = n + 10, colors = (1, 0, 0) )
        #print('####### hlines ##########')
        xmin = X[0]
        xmax = X[-1]
        #print(f, xmin, xmax)
        plt.hlines(y_hline, xmin, xmax, colors = color, linestyles = 'dotted' )
        
    # plot the line chart
    plt.plot(X, y, color = (0.5, 0, 0.5) )
    #plt.ylim(20,50)
    
    # save frame
    plt.savefig(fig_path)
    plt.close()
        

passo = 30
fig_paths = []
qtde_imgs = 100
for i in range(qtde_imgs):
    n_min = i
    n_max = n_min + passo 
    #print(passo, n_min, n_max)
    
    fig_path = os.path.join('TempImages', f'{i}.png')
    fig_paths.append(fig_path)
    create_frame(passo, n_min, n_max, fig_path=fig_path)
    #print(f'path({i}) = {fig_path}')
    
# build gif
with imageio.get_writer('visualization.gif', mode='I') as writer:
    for fig_path in fig_paths:
        image = imageio.imread(fig_path)
        writer.append_data(image)
        
# Remove files
#for fig_path in set(fig_paths):
#    os.remove(fig_path)
```
